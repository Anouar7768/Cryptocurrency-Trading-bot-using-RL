# Import of data and libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df_nupl_sent = pd.read_csv("../data/nupl_200_steps_with_sentiment.csv", index_col="Unnamed: 0")
df_nupl_without_sent = pd.read_csv("../data/nupl_200_steps_without_sentiment.csv", index_col="Unnamed: 0")
```

```python
df_nupl_sent.columns = ["NUPL_with_sentiment"]
df_nupl_without_sent.columns = ["NUPL_without_sentiment"]
```

```python
data = pd.read_csv("../data/all_data.csv")
```

```python
# Add the date column 
real_BNB = data.BNB[32:232]
real_BTC = data.BTC[32:232]
real_ETH = data.ETH[32:232]

date = data.Date[32:232]
```

```python
df_nupl = pd.DataFrame(np.hstack((np.hstack((np.array(df_nupl_sent.values), np.array(df_nupl_without_sent.values))), date)))
```

```python
df_nupl.columns = ["NUPL_with_sentiment", "NUPL_without_sentiment", "date"]
df_nupl = df_nupl.set_index('date')
```

```python
# Plot the three cryptos prices on the same graph
plt.plot(real_BNB, label='BNB')
plt.plot(real_BTC, label='BTC')
plt.plot(real_ETH, label='ETH')

# Add a legend to the plot
plt.legend()
```

# Plot the difference with and without sentiment

```python
ax = df_nupl.plot(kind='line', y=['NUPL_with_sentiment', 'NUPL_without_sentiment'], xlabel='day', ylabel='NUPL')

# Show the plot
plt.show()
```
