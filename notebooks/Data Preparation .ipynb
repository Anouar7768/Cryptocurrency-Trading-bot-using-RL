```python
def get_mean_value(df, crypto_name):
    df[f"market_value"] = (df['Open'] + df['Close'])/2
    df = df.reset_index()
    return df 
```

## Sentiment score dataset

# Concatenate all crypto data

```python
## Preidctions done with sentiment 
BNB = pd.read_csv("../data/predictions_with_sentiment_BNB.csv").to_dict() 
BTC = pd.read_csv("../data/predictions_with_sentiment_BTC.csv").to_dict() 
ETH = pd.read_csv("../data/predictions_with_sentiment_ETH.csv").to_dict() 
```

```python
all_data = {}
for crypto, data in zip(['BNB',"BTC",'ETH'], [BNB,BTC,ETH]):
    all_data[crypto] = data['Predicted_price_with_sentiment']

all_data['Date'] = BTC['Date']
```

```python
pd.DataFrame.from_dict(all_data).to_csv('../data/all_data_with_sentiment.csv',index=False)
```

```python
## Preidctions done without sentiment 
BNB = pd.read_csv("../data/predictions_without_sentiment_BNB.csv").to_dict() 
BTC = pd.read_csv("../data/predictions_without_sentiment_BTC.csv").to_dict() 
ETH = pd.read_csv("../data/predictions_without_sentiment_ETH.csv").to_dict() 
```

```python
all_data = {}
for crypto, data in zip(['BNB',"BTC",'ETH'], [BNB,BTC,ETH]):
    all_data[crypto] = data['Predicted_price_without_sentiment']

all_data['Date'] = BTC['Date']
```

```python
pd.DataFrame.from_dict(all_data).to_csv('../data/all_data_without_sentiment.csv',index=False)
```
